### Series scraper testing

```python
# functions
def date_range(start, end):
    r = (end+datetime.timedelta(days=1)-start).days
    return [datetime_to_str(start+datetime.timedelta(days=i)) for i in range(r)]
def datetime_to_str(dt):
    return dt.strftime("%Y-%m-%dT%M:%S")

def pickle_dates(startDate,endDate,market,grp_type,node):
    #check if file has already been created
    if not os.path.isfile(os.path.join(savedir,"LMP_%s_%s.pi" %(market,startDate))):
        data = mycaiso.get_lmp(False,startDate,endDate,market,grp_type,node,csv=True)
        data.to_pickle("%s/LMP_%s_%s.pi" %(savedir,market,startDate))
        return data
    else:
        pass
    
def get_summ_stats(date_str):

    # choose day
    date = "day_long_datasets/overgen_2014-%s" %date_str

    #load data
    data = pd.read_pickle(savedir+"/%s.pi" %date)

    node_cols = list(data.columns)
    for val in ['mean','min','max']:
        data[val]=getattr(data[node_cols],val)(axis=1)
    
    data['q_95'] = data[node_cols].quantile(.95,axis=1)
    data['q_05'] = data[node_cols].quantile(.05,axis=1)
    summ = data[['mean','min','max','q_05','q_95']]
    summ = summ.sortlevel(axis = 0)
    lmp_data = summ.loc(axis=0)[:,"LMP"]
    lmp_data.index = lmp_data.index.droplevel("LMP_TYPE")
    mcc_data = summ.loc(axis=0)[:,"MCC"]
    mcc_data.index = mcc_data.index.droplevel("LMP_TYPE")
    
    return lmp_data,mcc_data

def print_node_data(df_loc,lmp_type,overgen_only=False):
    total_year = pd.read_pickle(df_loc)
    lmp_year = total_year[total_year['LMP_TYPE']==lmp_type]
    del lmp_year['LMP_TYPE']
    
    if overgen_only==True:
        lmp_year = lmp_year[lmp_year['i_overgen']==True]
    try: del lmp_year['i_overgen']
    except: pass

    instances_below_0 = lmp_year[lmp_year<0].count()
    lmp_early = lmp_year[:'2014-03-31']
    instances_below_floor_early = lmp_early[lmp_early<-30].count()
    lmp_late = lmp_year['2014-04-01':]
    instances_below_floor_late = lmp_late[lmp_late<-100].count()
    instances_below_floor = instances_below_floor_early.add(instances_below_floor_late)
    instances_above_ceil = lmp_year[lmp_year>1000].count()

    nodes_below_0 = sum(instances_below_0 > 0)
    nodes_below_floor = sum(instances_below_floor > 0)
    nodes_above_ceil = sum(instances_above_ceil > 0)

    nhours_below_0 = instances_below_0.sum() * 365.25 / 12
    nhours_below_floor = instances_below_floor.sum() *265.25 / 12
    nhours_above_ceil = instances_above_ceil.sum() *265.25 / 12

    print "# of nodes below $0:", nodes_below_0
    print "# of node hours below $0:", nhours_below_0
    print "% of node hours below $0:", 100 * nhours_below_0 * (12/365.25) / float(lmp_year.count().sum())
    print "# of nodes below $-30:", nodes_below_floor
    print "# of node hours below $-30:", nhours_below_floor
    print "% of node hours below $-30:", 100 * nhours_below_floor * (12/365.25) / float(lmp_year.count().sum())
    print "# of nodes above $1000:", nodes_above_ceil
    print "# of node hours above $1000:", nhours_above_ceil
    print "% of node hours above $1000:", 100 * nhours_above_ceil * (12/365.25) / float(lmp_year.count().sum())
```

```python
#Pick date to pull LMP data for
date_list = []
end_date_list = []
for date in ['04-22']:
    
    # Prepare to gather data from day of event, plus 1 week before and 1 week after
    date = datetime.datetime.strptime("2014-"+date,"%Y-%m-%d")
    plus_minus = datetime.timedelta(days=7)
    surrounding_dates = [date-plus_minus, date, date+plus_minus]
    
    for j in surrounding_dates:
        date_list_temp = []
        end_date_list_temp = []
        for i in range(25):
            next_dt = j+datetime.timedelta(hours=i)
            next_dt = next_dt.strftime("%Y-%m-%dT%H:%M")
            date_list_temp.append(next_dt)

        end_date_list_temp = date_list_temp[1:]
        date_list_temp = date_list_temp[:-1]
        date_list = date_list + date_list_temp
        end_date_list = end_date_list + end_date_list_temp
num_dates = len(date_list)

# generate lists of input strings for parallelized scraper function
market = ['RT5M']*num_dates
grp_type = ['ALL']*num_dates
node = ['ALL']*num_dates

# pull data
results = lv.map(pickle_dates,date_list,end_date_list,market,grp_type,node)
```

### Append together a day's worth of RTM data

```python
plus_minus = timedelta(days=7)
overgen_dates = True

if overgen_dates: dates_to_loop = overgen_dates_2014
else: dates_to_loop = [i+"-01" for i in monthly_dates]
for date in dates_to_loop:
    print date
    date_full = "2014-%s" %date
    
    dt = datetime.strptime(date_full,"%Y-%m-%d")
    if overgen_dates:
        # Prepare to gather data from day of event, plus 1 week before and 1 week after
        surrounding_dates = [dt-plus_minus, dt, dt+plus_minus]
    else: surrounding_dates = [dt]

    files = []
    for i in surrounding_dates:
        dt_str = i.strftime("%Y-%m-%d")
        files = files + glob(os.path.join(savedir,"hour_long_datasets","*%s*" %dt_str))
    assert len(files)==24*len(surrounding_dates)
    
    day_long_df = pd.DataFrame()
    for i in files:
        day_long_df = day_long_df.append(pd.read_pickle(i))

    # convert gmt to pst
    day_long_df.index = day_long_df.index.droplevel('INTERVALENDTIME_GMT')
    day_long_df = day_long_df.reset_index(level='LMP_TYPE',drop=False)
    day_long_df.index.tz = 'GMT'
    day_long_df.index = day_long_df.index.tz_convert('US/Pacific')
    day_long_df.index.name = 'time'
    day_long_df = day_long_df.reset_index(drop=False)
    day_long_df = day_long_df.set_index(['time','LMP_TYPE'])
    
    # save appended dataframe
    if overgen_dates: save_prefix = "overgen_"
    else: save_prefix = ""
    day_long_df.to_pickle(savedir + "/day_long_datasets/" + save_prefix + date_full + ".pi")
```

#### Plotting

#### mean, min, max LMP for all 2014 days with reported overgen events (through April)

```python
ncols = 2
nrows = int(ceil(len(overgen_dates_2014)/ncols))
fig1, axes = plt.subplots(nrows=nrows, ncols=ncols,figsize=(12,12))
axes_list = [[i,j] for i in range(nrows) for j in range(ncols)]

for j,i in enumerate(overgen_dates_2014):
    lmp_data = get_summ_stats(i)[0]
    
    row_num = axes_list[j][0]
    col_num = axes_list[j][1]
    plot_tmp = lmp_data['2014-%s' %i][['min','mean','max']].plot(ax=axes[row_num,col_num])
    ymin,ymax = axes[row_num,col_num].get_ylim()
    plot_tmp.vlines(x=overgen_hours_2014[i],ymin=ymin,ymax=ymax,color='k')
    axes[row_num,col_num].set_title(i + "-14")
    axes[row_num,col_num].set_ylabel("LMP ($)")

#save
fig1.savefig(os.path.join(figdir,"overgen_days_rtm_lmp_min_mean_max.png"))
```
