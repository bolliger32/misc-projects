```python
# functions
def date_range(start, end):
    r = (end+datetime.timedelta(days=1)-start).days
    return [datetime_to_str(start+datetime.timedelta(days=i)) for i in range(r)]
def datetime_to_str(dt):
    return dt.strftime("%Y-%m-%dT%M:%S")

def pickle_dates(startDate,endDate,market,grp_type,node):
    #check if file has already been created
    if not os.path.isfile(os.path.join(savedir,"LMP_%s_%s.pi" %(market,startDate))):
        data = mycaiso.get_lmp(False,startDate,endDate,market,grp_type,node,csv=True)
        data.to_pickle("%s/LMP_%s_%s.pi" %(savedir,market,startDate))
        return data
    else:
        pass
```

### Test parallel downloader function

```python
# Start ipython engines
from IPython import parallel
clients = parallel.Client()
dv = clients.direct_view()
lv = clients.load_balanced_view()
```

```python
# Load common variables onto engines
dv.execute("""
import pandas as pd
import os
homedir = "/Users/ianbolliger/Documents/git/pyiso"
savedir = "/Users/ianbolliger/Documents/LMP/RTM"
os.chdir(homedir)
from pyiso import caiso
mycaiso = caiso.CAISOClient()
""")
```

```python
#Pick date to pull LMP data for
date_list = []
end_date_list = []
for date in ['01-23','04-12','04-25','04-26','04-27']:
    
    # Prepare to gather data from day of event, plus 1 week before and 1 week after
    date = datetime.datetime.strptime("2014-"+date,"%Y-%m-%d")
    plus_minus = datetime.timedelta(days=7)
    surrounding_dates = [date-plus_minus, date, date+plus_minus]
    
    for j in surrounding_dates:
        date_list_temp = []
        end_date_list_temp = []
        for i in range(25):
            next_dt = j+datetime.timedelta(hours=i)
            next_dt = next_dt.strftime("%Y-%m-%dT%H:%M")
            date_list_temp.append(next_dt)

        end_date_list_temp = date_list_temp[1:]
        date_list_temp = date_list_temp[:-1]
        date_list = date_list + date_list_temp
        end_date_list = end_date_list + end_date_list_temp
num_dates = len(date_list)

# generate lists of input strings for parallelized scraper function
market = ['RT5M']*num_dates
grp_type = ['ALL']*num_dates
node = ['ALL']*num_dates

# pull data
results = lv.map(pickle_dates,date_list,end_date_list,market,grp_type,node)
```

### Append together a day's worth of RTM data

```python
results.ready()
```

```python
plus_minus = datetime.timedelta(days=7)
for date in ['01-23','04-12','04-25','04-26','04-27']:
    date_full = "2014-%s" %date
    # Prepare to gather data from day of event, plus 1 week before and 1 week after
    dt = datetime.datetime.strptime(date_full,"%Y-%m-%d")
    surrounding_dates = [dt-plus_minus, dt, dt+plus_minus]

    files = []
    for i in surrounding_dates:
        dt_str = i.strftime("%Y-%m-%d")
        files = files + glob(savedir+"/*%s*" %dt_str)
    print date, len(files)
    assert len(files)==72
    
    day_long_df = pd.DataFrame()
    for i in files:
        day_long_df = day_long_df.append(pd.read_pickle(i))

    # save appended dataframe
    day_long_df.to_pickle(savedir + "/day_long_datasets/overgen_" + date_full + ".pi")
```

### Append together once-a-month data across a year

```python
year_long_df = pd.DataFrame()
date = "2014"
files = glob(savedir+"/day_long_datasets/*%s*" %date)
assert len(files)==12
for i in files:
    year_long_df = year_long_df.append(pd.read_pickle(i))
    
# save appended dataframe
year_long_df.to_pickle(savedir + "/year_long_datasets/" + date + ".pi")
```

### Export formatted csvs for electricitymapper

```python
node_names = pd.read_csv(elec_mapper_sample_csv,index_col='name').index.values

date = "2014-02-01"
data = pd.read_pickle(savedir+"/day_long_datasets/%s.pi" %date)
lmp_data = [data.loc(axis=0)[:,:,"LMP"],"lmp"]
mcc_data = [data.loc(axis=0)[:,:,"MCC"],"mcc"]
for i in [lmp_data,mcc_data]:
    i[0] = i[0].reset_index(level=(1,2),drop=True)
    i[0] = i[0].T
    i[0].index.name='name'
    i[0] = i[0].reindex(node_names)
    i[0].to_csv("/Users/ianbolliger/LMP/emapper_csvs/%s_%s" % (i[1],date))
    
```
