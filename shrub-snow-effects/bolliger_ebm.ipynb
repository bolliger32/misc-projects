```python
## parameter input

#snow parameters
A = .5 # albedo
T_s = 273.15 # during snowmelt phase, assume snow is at 0 deg (probably not valid at night during melt-freeze cycle)
epsilon_s = .98 # from Tarboton et al 1995

#solar/atmospheric parameters
daylight_hrs = 20.5 # hours at which the sun is over the horizon (~19 for 65N in May)
gamma = 6/7. #fraction direct light for clear skies
lat = 65 # latitude
day = 94 # April 1
a_s = .25 # atmospheric transmissivity on cloudy day
b_s = .5 # additional transmissivity on clear day
clf = 0 # cloud fraction
e_a = 400 #vapor pressure (Pa) - number taken from NCEP reanalysis of mean monthly vapor pressure at 2m - April average across continental arctic basins
T_a_day = 285.15 #above canopy air-temp during day
T_a_night = T_a_day - 10 #above canopy air-temp during night
epsilon_a_cls = 1.08 * (1 - exp(-((e_a/100.)**(T_a/2016)))) # clear sky emissivity

#canopy parameters
G = .5 # leaf orientation factor
a = .5 # leaf scattering coefficient
LF = 1 # canopy level LAI
epsilon_c = .98 # from Tarboton et al 1995
T_c_day = T_a_day # assuming canopy is at ambient temperature
T_c_night = T_a_night # again assume canopy stays at ambient temp


## intermediate parameter calculations
#canopy
k = sqrt(1-a)
beta_inf = (1-k)/(1+k)
Kb = G / cos(zen)

#solar/atmospheric
night_hrs = 24 - daylight_hrs
Q_TOA = daily_insolation(65,155) * 24 / daylight_hrs # average daily insolation restricted to daylight hours (should be sinusoidal)
AT = a_s + (1-clf)*b_s # effetive atmospheric transmissivity
AT_b = gamma*(a_s+b_s)*(1-clf)
AT_d = AT - AT_b
Q0 = Q_TOA * AT
Qb = Q0 * (AT_b/AT)
Qd = Q0 * (AT_d/AT)
    
epsilon_a = clf + (1-clf)*epsilon_a_cls

# other parameters
zen_deg = rad2deg(zen)
```

```python
def calc_net_fluxes_no_canopy(loc,night=False):
    if night:
        T_a = T_a_night
    else:
        T_a = T_a_day

        
    ## LW
    f_1i = epsilon_s
    f_3i = 1-epsilon_s
    f_1e = 0
    f_3e = 1

    
    ## Net radiative fluxes

    #SW
    if night:
        Q_sns = 0
        Q_rns = 0   
    else:
        Q_sns = (1-A) * Q0
        Q_rns = A * Q0

    #LW
    Q_li = epsilon_a * sigma * T_a**4
    Q_le = epsilon_s * sigma * T_s**4
    Q_snl = f_1i*Q_li - Q_le
    Q_rnl = f_3i*Q_li + Q_le - Q_li

    #net
    Q_sn = Q_sns + Q_snl
    Q_rn = Q_rns + Q_rnl
    
    result = {"Q_sns":Q_sns,
            "Q_rns":Q_rns,
            "Q_snl":Q_snl,
            "Q_rnl":Q_rnl,
            "Q_sn":Q_sn,
            "Q_rn":Q_rn}
    
    if loc == None:
        return result
    else:
        return result[loc]
    

def calc_net_fluxes(LF,zen,loc,night=False):
    if night:
        T_a = T_a_night
        T_c = T_c_night
        zen = 0
    else:
        zen_deg = rad2deg(zen)
        T_a = T_a_day
        T_c = T_c_day
    Kb = G / cos(zen)


    def f1(tau):
        return (1-A)*tau / (1 - A*beta_d_finite*(1 - t_d_finite))

    def f3(tau,beta):
        return ((1-A)*beta + A*tau*t_d_finite) / (1 - A*beta_d_finite*(1-t_d_finite))

    def f_net(action,type):
        """To determine fraction of incident solar radiation reflected, absorbed, and transmitted by canopy
        after multiple scattering within finite canopy and multiple reflections between canopy and surface

        Inputs:
            action (str): 'refl','absrb','transm'
            type (str): 'b' (direct), 'd' (diffuse)

        Outputs:
            return: float
        """
        if type == 'b':
            tau = t_b_finite
            beta = beta_b_finite
        elif type == 'd':
            tau = t_d_finite
            beta = beta_d_finite
        else:
            raise NameError()

        if action == 'refl':
            return f3(tau,beta)
        elif action == 'transm':
            return f1(tau)
        elif action == 'absrb':
            return 1 - f3(tau,beta) - f1(tau)
        else:
            raise NameError()

    ## define net transmissivity, reflectance, and absorbtion of canopy using Mahat and Tarboton model
    #SW

    # direct
    t_b_beers = exp(-Kb*LF) #Beer's law transmissivity of direct light
    t_b_inf = exp(-k*Kb*LF) #Direct transmissivity with semi-infinite multiple scattering property
    t_b_finite = t_b_inf * ((1-(beta_inf**2)))/(1-(beta_inf**2)*(t_b_inf**2)) #Direct transmissivity with finite canopy and scattering
    beta_b_finite = (beta_inf*(1-(t_b_inf**2))) / (1 - (beta_inf**2)*(t_b_inf**2))

    # diffuse
    ##t_d_beers not defined in Mahat and Tarboton, but can simply assume k = 1 (a=0) and use semi-infinite transmissivity
    t_d_beers = ones_like(t_b_inf) * (1-G*LF)*exp(-G*LF) + (G*LF)**2 * expn(1,G*LF)
    t_d_inf = ones_like(t_b_inf) * (1-k*G*LF)*exp(-k*G*LF) + (k*G*LF)**2 * expn(1,k*G*LF)
    t_d_finite = t_d_inf * ((1-(beta_inf**2)))/(1-(beta_inf**2)*(t_d_inf**2))
    beta_d_finite = (beta_inf*(1-(t_d_inf**2))) / (1 - (beta_inf**2)*(t_d_inf**2)) #diffuse reflectance

    # final model (including multiple reflections from surface)
    transm_b = f_net('transm','b')
    transm_d = f_net('transm','d')
    refl_b = f_net('refl','b')
    refl_d = f_net('refl','d')
    absrb_b = f_net('absrb','b')
    absrb_d = f_net('absrb','d')



    ## Net radiative fluxes

    #SW
    if night:
        Q_sns = 0
        Q_cns = 0
        Q_rns = 0   
    else:
        Q_sns = transm_b * Qb + transm_d * Qd
        Q_cns = absrb_b * Qb + absrb_d * Qd
        Q_rns = refl_b * Qb + refl_d * Qd
        
        
    #LW
    f_1i = epsilon_s*t_d_finite
    f_2i = (1-t_d_finite) * epsilon_c + t_d_finite * (1-epsilon_s)
    f_3i = (1-t_d_finite) * (1 - epsilon_c)
    f_1e = (1-t_d_finite) * (1-epsilon_c)
    f_2e = (1-t_d_finite) * epsilon_c
    f_3e = t_d_finite
    f_1c = epsilon_s
    f_2c = (1-t_d_finite) * (1-epsilon_s)*epsilon_c
    f_3c = t_d_finite * (1-epsilon_s)
    

    Q_li = epsilon_a * sigma * T_a**4
    Q_lc = epsilon_c * sigma * T_c**4 * (1-t_d_finite)
    Q_le = epsilon_s * sigma * T_s**4
    Q_snl = f_1i*Q_li + f_1e*Q_le + f_1c*Q_lc - Q_le
    Q_cnl = f_2i*Q_li + f_2e*Q_le + f_2c*Q_lc - 2*Q_lc
    Q_rnl = f_3i*Q_li + f_3e*Q_le + f_3c*Q_lc - Q_lc

    #net
    Q_sn = Q_sns + Q_snl
    Q_cn = Q_cns + Q_cnl
    Q_rn = Q_rns + Q_rnl

    result = {"Q_sns":Q_sns,
            "Q_cns":Q_cns,
            "Q_rns":Q_rns,
            "Q_snl":Q_snl,
            "Q_cnl":Q_cnl,
            "Q_rnl":Q_rnl,
            "Q_sn":Q_sn,
            "Q_cn":Q_cn,
            "Q_rn":Q_rn,
            "t_b_beers":t_b_beers,
            "t_d_beers":t_d_beers,
            "t_b_finite":t_b_finite,
            "t_d_finite":t_d_finite,
            "Qb":Qb,
            "Qd":Qd}

    if loc == None:
        return result
    else:
        return result[loc]
```

```python
## Calc daily flux given max zenith angle and sun elevation)
def elev_cur(t_sec,daylight_dur_hrs,elev_max):
    result = arctan(tan(elev_max)*sin(t_sec*pi/(3600.*daylight_dur_hrs)))
    return result

def flux_by_time(t_sec, daylight_dur_hrs, elev_max, LF, loc):
    zen = (pi/2) - elev_cur(t_sec, daylight_dur_hrs, elev_max)
    return calc_net_fluxes(LF, zen, loc)
@vectorize
def integrate_daily_flux(daylight_dur_hrs,elev_max, LF,loc,canopy=True):
    night_dur_hrs = 24 - daylight_dur_hrs
    if canopy:
        joules_per_day = integrate.quad(flux_by_time, 0, daylight_dur_hrs*3600, args = (daylight_dur_hrs,elev_max, LF, loc))[0]
        joules_per_night = calc_net_fluxes(LF,0,loc,night=True)* night_dur_hrs*3600
    else:
        joules_per_day = calc_net_fluxes_no_canopy(loc,night=False)*daylight_dur_hrs*3600
        joules_per_night = calc_net_fluxes_no_canopy(loc,night=True)*night_dur_hrs*3600
    net_daily_flux = joules_per_day + joules_per_night
    net_daily_flux = net_daily_flux / (3.6e6)
    return net_daily_flux
```
